```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
plt.style.use('dark_background')
```

```python
df = pd.read_csv('./files/price_prediction.csv')
df.head()
```

</p>Here we will only work with quantitative featrues</p>

<h3>Different corrleation methods</h3>

```python
df = df.drop(columns=['Id'])
```

```python
columns = df.corr().columns
```

```python
f = pd.melt(df[columns], id_vars='SalePrice', value_vars=columns)
g = sns.FacetGrid(f, col='variable', col_wrap=3, sharex=False, sharey=False, height=3)
g.map(sns.scatterplot, 'value', 'SalePrice')
plt.show()
```

<p>For this test we will take some selective features which shows plots where best fit line can be somthing other than linear line</p>

```python
features = ['OverallQual', 'OverallCond', 'YearBuilt', 'TotalBsmtSF', 'GarageYrBlt', 'SalePrice']

f = pd.melt(df[features], id_vars='SalePrice', value_vars=features)
g = sns.FacetGrid(f, col='variable', col_wrap=3, sharex=False, sharey=False)
g.map(sns.scatterplot, 'value', 'SalePrice')
plt.show()
```

```python
corr_pearson = df[features].corr()
sns.heatmap(corr_pearson, annot=True, fmt='.2f')
plt.show()
```

```python
df['SalePrice'].corr(df['OverallCond'])
```

```python
corr = df['SalePrice'].corr(df['OverallCond'])

print(f'The corr between OverallQual and Saleprice in pearson method is {corr}')
```

```python
corr = df['SalePrice'].corr(df['OverallCond'], 'spearman')

print(f'The corr between OverallQual and Saleprice in pearson method is {corr}')
```

<p><b>Pearson</b> method is affected by outliers and can not pickup no linear relation correctly but <b>Spearman</b> method does not have this problem as it assigns ranks and so it is not affected by outliers and so it can pickup no linearrelation much more correctly</p>

```python
corr_spearman = df[features].corr('spearman')
sns.heatmap(corr_spearman, annot=True, fmt='.2f')
plt.show()
```

```python
fig, axis = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

func = lambda x: round(x, 2)

ax = sns.barplot(x=corr_pearson['SalePrice'].apply(func), y=corr_pearson['SalePrice'].index, ax=axis[0])
ax.bar_label(ax.containers[0])

ax = sns.barplot(x=corr_spearman['SalePrice'].apply(func), y=corr_spearman['SalePrice'].index, ax=axis[1])
ax.bar_label(ax.containers[0])

plt.tight_layout()
plt.show()
```

<p>You can see that by using pearson method we see that the releation between the features has increased greatly as spearman method is not affected that much by oulayers. Also spearman method can pickup non linear relationship which can be seen from 
<b>YearBuilt and GarageYrBlt</b> features</p>

<h3>Transformation</h3>

<p>Now we will see the affect of transformation on YearBuilt and GarageYrBlt and TotalBsmt</p>

```python
x_base = df['YearBuilt']
y_base = df['SalePrice']
sns.scatterplot(x=x_base, y=y_base)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')}")
```

<p>You can see that the relationship between YearBuilt and SalePrice is clearly not linear it looks more like a exponensitial function. So if you try to fit a linear line here the line would not be perfect</p>

```python
# so if take the ln on of YearBuilt
x = df['YearBuilt'].apply(lambda x: np.log(x))
y = df['SalePrice']
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# taking ln of saleprice
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: np.log(x))
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# taking ln twice of salerprice
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: np.log(np.log(x)))
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# taking squre root of SalePrice
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: np.sqrt(x))
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# taking multiple square root of SalePrice and seeing pearson corr increases with more sqrt(as it appoches previously taken ln
# value the corr reatlion becomes the same)

x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: np.sqrt(np.sqrt(np.sqrt(x))))
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# square of saleprice
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: x ** 2)
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

```python
# cube of SalePrice
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: x ** 3)
sns.scatterplot(x=x, y=y)
plt.show()

print(f"The initial pearson corr is {x_base.corr(y_base)} and spearman corr {x_base.corr(y_base, 'spearman')} and new pearson\
 corr {x.corr(y)} and spearman corr is {x.corr(y, 'spearman')}")
```

<h4>From all these tests we can again verify the truth that spearman method can easily get the relationship between two features even if they are not in a linear relation. As from the test above you can tell that no matter what transformation we apply the the spearman correlation does not change as the plot getters closer to a linear relationship the pearson corr changes for the better and for the ln transformation we get the easy and best relation </h4>

<h3>Using Linear model with both transformed and non transformed data</h3>

<p>For simplicity we will only use the YearBuilt feature to predict price</p>

```python
x = df['YearBuilt']
y = df['SalePrice']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=20, test_size=0.2)
```

```python
x_train.values.reshape(-1, 1)
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()

x_train = scaler_x.fit_transform(x_train.values.reshape(-1, 1)).flatten()
x_test = scaler_x.transform(x_test.values.reshape(-1, 1)).flatten()

y_train = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()
```

```python
reg = LinearRegression()
reg.fit(x_train.reshape(-1, 1), y_train)
```

```python
y_pred = reg.predict(x_test.reshape(-1, 1))
```

```python
reg.score(x_test.reshape(-1, 1), y_test)
```

<p>If we use transformed SalePrice</p>

```python
x = df['YearBuilt']
y = df['SalePrice'].apply(lambda x: np.log(x))

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=20, test_size=0.2)
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()

x_train = scaler_x.fit_transform(x_train.values.reshape(-1, 1)).flatten()
x_test = scaler_x.transform(x_test.values.reshape(-1, 1)).flatten()

y_train = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_test = scaler_y.transform(y_test.values.reshape(-1, 1)).flatten()
```

```python
reg = LinearRegression()
reg.fit(x_train.reshape(-1, 1), y_train)
```

```python
reg.score(x_test.reshape(-1, 1), y_test)
```

<h3>As you can see after transforming the SalePrice the accuracy score although very less but has definetly improved. Here we only use one feature for simplicty but in future we will use the whole dataset for predicting saleprice</h3>
