<h3>Removig Trends</h3>

```python
y = co.copy()


X = DeterministicProcess(
    index = y.index,
    order = 1,
    constant = True,
    drop = True
)

x = X.in_sample()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(x, y)
```

```python
y_pred = reg.predict(x)

plt.plot(x.index, y_pred)

plt.xticks(rotation='vertical')

plt.show()
```

```python
y_residual = y - y_pred


y_residual.plot(color='0.25', markerfacecolor='0.15', style='.-', figsize=(12, 5))

plt.xticks(rotation='vertical')


co.plot(color='0.25', markerfacecolor='0.15', style='.-', figsize=(12, 5))

plt.show()
```

```python
resd_rolling = y_residual.rolling(
    window=12,
    min_periods = 6,
    center =True,
).mean()

resd_rolling.head()
```

```python
_, ax = plt.subplots(figsize=(12, 5))

# ax = y.plot(color='0.25', markerfacecolor='0.15', style='.-', ax=ax)
ax = resd_rolling.plot(ax=ax)
```
